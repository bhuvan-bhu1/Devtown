```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films'
req = requests.get(url)
req
```

```python
soup = BeautifulSoup(req.content,"html.parser")
soup.prettify()
```

```python
array = []
for i in soup.find_all("td"):
    array.append(str(i))
array[0]
```

```python
re.sub('^<td>.*">|<.*>',"",array[4])
```

```python
array[1]
```

```python
re.sub('^<.*">|</a.*d>',"",array[5])
```

```python
array[2]
```

```python
re.sub('^<td>|</td>',"",array[6])
```

```python
array[7]
```

```python
re.sub('^<td>|\n.*d>',"",array[7])
```

```python
array[10]
```

```python
films = []
years = []
awards = []
nominis = []
```

```python
len(array)
```

```python
i = 0
while i != (len(array) - 2):
    film = re.sub('^<td>.*">|<.*>',"",array[i])
    films.append(film)
    i += 1
    year = re.sub('^<.*">|</a.*d>',"",array[i])
    years.append(year)
    i += 1
    award = re.sub('^<td>|</td>',"",array[i])
    awards.append(award)
    i += 1
    nomini = re.sub('^<td>|\n.*d>',"",array[i])
    nominis.append(nomini)
    i += 1
```

```python
len(films)
```

```python
len(years)
```

```python
len(awards)
```

```python
len(nominis)
```

```python
data = {"Film" : films ,"Year" : years ,"Awards" : awards ,"Nominations" : nominis}
len(data["Film"])
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.tail()
```

```python
df.drop(1348,inplace=True)
df.drop(1347,inplace=True)
```

```python
df.tail()
```

```python
df.to_csv("Top_movies.csv")
```
