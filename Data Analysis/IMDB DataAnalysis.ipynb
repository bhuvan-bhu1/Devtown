```python
import pandas as pd
import numpy as np
import sqlite3
import matplotlib.pyplot as plt
```

```python
data = sqlite3.connect("movie.sqlite")
cursor = data.cursor()
```

```python
cursor.execute("pragma table_info('movies')")
table1 = cursor.fetchall()
table1
```

```python
cursor.execute("pragma table_info('directors')")
table2 = cursor.fetchall()
table2
```

```python
cursor.execute('select * from movies')
table1_info = cursor.fetchall()
table1_info = pd.DataFrame(table1_info,columns=["id","original_title","budget","popularity","release_date","revenue","title","vote_average","vote_count","overview","tagline","uid","director_id"])
table1_info.head()
```

```python
cursor.execute('select * from directors')
table2_info = cursor.fetchall()
table2_info = pd.DataFrame(table2_info,columns=["name","id","gender","uid","department"])
table2_info.head()
```

```python
#movies present in the Imdb Database
cursor.execute('select count(title) from movies')
total_movies = cursor.fetchall()
total_movies
```

```python
#get the total info of James cameron ,Luc Besson ,John Woo
cursor.execute('select * from directors where name in ("James Cameron","Luc Besson","John Woo")')
content_of_particular_director = cursor.fetchall()
content_of_particular_director
```

```python
#All the info whose name start with Steven 
cursor.execute('select * from directors where name like "Steven%"')
start_with = cursor.fetchall()
start_with
```

```python
#Total number of Female Directors in Movies Database
cursor.execute('select count(*) from directors where gender = "1"')
total_female_directors = cursor.fetchall()
total_female_directors[0][0]
```

```python
#Name of the Tenth Women Director In Movie Database
cursor.execute('select name from directors where gender = "1" limit 10')
first_ten_female_directors = cursor.fetchall()
first_ten_female_directors[9][0]
```

```python
#List of Top 3 Popular Movies in Movie Database
cursor.execute('select title from movies order by popularity desc limit 3')
top_three_popular_movies = cursor.fetchall()
top_three_popular_movies
```

```python
#List of Top 3 Bankable Movies in Movie Database
cursor.execute('select title from movies order by revenue desc limit 3')
top_three_revenue_movies = cursor.fetchall()
top_three_revenue_movies
```

```python
#Most Awarded movie from jan 1st 2000 in Movie Database
cursor.execute('select title from movies where release_date > "2000-01-01" order by vote_average limit 1 ')
most_awarded = cursor.fetchall()
most_awarded[0][0]
```

```python
#List of Movies directed by Brenda Chapman from Movie.Database
cursor.execute('select title from movies inner join directors on movies.director_id = directors.id where name = "Brenda Chapman"')
movies_directed_by = cursor.fetchall()
movies_directed_by[0][0]
```

```python
#Name of the director who made most number of movies in Movie Database
cursor.execute('select name from directors inner join movies on directors.id = movies.director_id group by director_id order by count(name) desc limit 1')
name_of_the_director_who_made_most_movies = cursor.fetchall()
name_of_the_director_who_made_most_movies[0][0]
```

```python
#Name of the director who is most rich or profitable of movies in Movie Database
cursor.execute('select name from directors inner join movies on directors.id = movies.director_id group by director_id order by sum(revenue) desc limit 1')
name_of_the_director_who_made_most_bankable = cursor.fetchall()
name_of_the_director_who_made_most_bankable[0][0]
```

```python
#List of Top 10 Highest budget Making movies in Movie Database
cursor.execute('select title from movies order by budget desc limit 10')
top_ten_budget_making_movies = cursor.fetchall()
top_ten_budget_making_movies
```

```python
#List of Top 10 Highest Revenue Making movies in Movie Database
cursor.execute('select title from movies order by revenue desc limit 10')
top_ten_revenue_making_movies = cursor.fetchall()
top_ten_revenue_making_movies
```

```python
#Top 10 highest voted movie info in IMDB Movie Database
cursor.execute('select * from movies order by vote_average desc limit 10')
highest_vote_average = cursor.fetchall()
highest_vote_average = pd.DataFrame(highest_vote_average,columns=["id","original_title","budget","popularity","release_date","revenue","title","vote_average","vote_count","overview","tagline","uid","director_id"])
highest_vote_average
```

```python
#Info about total number of movies and revenue
cursor.execute('select name,count(title),sum(revenue) from directors inner join movies on movies.director_id = directors.id group by director_id order by sum(revenue) desc')
director_movie = cursor.fetchall()
director_movie = pd.DataFrame(director_movie,columns=["Name" ,"Total_Movies","Revenue"])
director_movie.head()
```
