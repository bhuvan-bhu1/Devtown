```python
def call(arr,n):
    if n%2 == 0:
        p = n//2
    else:
        p = n//2 + 1
    final = []
    arr.sort()
    a = arr[:p]
    b = arr[p:]
    if len(a) >= len(b):
        for i in range(len(a)):
            try:
                final.append(a[i])
                final.append(b[i])
            except:
                final.append(a[i])
    else:
        for i in range(len(b)):
            try:
                final.append(a[i])
                final.append(b[i])
            except:
                final.append(b[i])
    if n%2 == 0:
        return final
    else:
        final.pop()
        return final
```
